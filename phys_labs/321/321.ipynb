### Исследование RLC-цепочки
Для R = 0 Ом и R = 100 Ом построим па одном графике зависимость $\frac{|\psi|}{\pi}$ от $\frac{\nu}{\nu_0}$

```python
gr3 = pd.read_csv('3.csv')
sns.scatterplot(data=gr3, x='nu0', y='psi_pi0')
sns.scatterplot(data=gr3, x='nu100', y='psi_pi100')
plt.xlabel(r'$\frac{|\psi|}{\pi}$')
plt.ylabel(r'$\frac{\nu}{\nu_0}$');
```

Обозначим за $2\Delta\nu$ ширину графика при $|\psi| = \frac{\pi}{4}$.

$$ 2\Delta_{R=0} = (0.18 \pm 0.06)\nu_0 $$

$$ 2\Delta_{R=100} = (0.56 \pm 0.15)\nu_0 $$

Добротность рассичтывается по формуле

$$ Q = \frac{\nu_0}{2\Delta\nu} $$

$$ Q_{э, R=0} = 6 \pm 2 $$

$$ Q_{э, R=100} = 1.8 \pm 0.5 $$

Рассчитаем добротность из параметров схемы.

$$ Q_{p, R=0} = 5.9 $$

$$ Q_{p, R=100} = 2.1 $$

Видно, что рассчетная добротность совпадает с измеренной в пределах погрешности.
