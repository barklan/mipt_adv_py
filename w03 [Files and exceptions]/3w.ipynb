### [pickle](https://docs.python.org/3/library/pickle.html) — Python object serialization

```python
import pickle
emp = {1: "A", 2: "B", 3: "C", 4: "D", 5: "E"}

# pickle this dict
with open('Emp.pickle', 'wb') as pickling_on:  # wb opens for writing in binary mode
    pickle.dump(emp, pickling_on)
```

```python
# unpickle this dict
with open('Emp.pickle', 'rb') as pickle_off:
    emp = pickle.load(pickle_off)
    print(emp)
```

## Объяснение про super()

[Python's super() documentation](https://docs.python.org/3/library/functions.html#super)

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * self.length + 2 * self.width

# Here we declare that the Square class inherits from the Rectangle class
class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)
```

Here, you’ve used **super()** to call the **\_\_init\_\_()** of the Rectangle class, allowing you to use it in the Square class without repeating code. Below, the core functionality remains after making changes:

```python
square = Square(4)
square.area()
```
